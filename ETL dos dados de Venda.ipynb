#Bibliotecas

```python
# Importando as bibliotecas necessários
import pandas as pd
import matplotlib.pyplot as plt
```

#Extração dos Dados

```python
clientes = pd.read_excel('/content/Dimensoes_DadosModelagem.xlsx', sheet_name='Clientes') #sheet_name me permite escolher a aba da planilha
func = pd.read_excel('/content/Dimensoes_DadosModelagem.xlsx', sheet_name='Funcionarios')
escri = pd.read_excel('/content/Dimensoes_DadosModelagem.xlsx', sheet_name='Escritorios')
categoria = pd.read_excel('/content/Dimensoes_DadosModelagem.xlsx', sheet_name='Categoria')
produtos = pd.read_excel('/content/Dimensoes_DadosModelagem.xlsx', sheet_name='Produtos')
fornec = pd.read_excel('/content/Dimensoes_DadosModelagem.xlsx', sheet_name='Fornecedores')
cabecalho = pd.read_table('/content/FatoCabecalho_DadosModelagem.txt')
detalhes = pd.read_csv('/content/FatoDetalhes_DadosModelagem.csv', sep=';') #sep mostra o separador do txt
```

#Preparação do data frame

```python
fun2 = func.merge(escri,
                  left_on = 'Escritorio',   
                  right_on = 'Escritorios',  
                  how = 'left')

print("Número de linhas e colunas:",fun2.shape)
fun2.head()
```

```python
# Removendo uma das colunas de escritorio
fun2.drop(columns = ['Escritorios'], inplace = True)

print("Número de linhas e colunas:",fun2.shape)
fun2.head()
```

```python
# Join com a tabela de produtos e categoria
produtos1 = produtos.merge(categoria,
                  left_on = 'CategoriaID',   
                  right_on = 'CategoriaID',  
                  how = 'left') 

# Join com a nova tabela de produtos e fornecedor
produtos2 = produtos1.merge(fornec,
                  left_on = 'FornecedorID',   
                  right_on = 'FornecedorID',  
                  how = 'left') 

print("Número de linhas e colunas:",produtos2.shape)
produtos2.head()
```

```python
# Join com a tabela detalhes e produtos
data1 = detalhes.merge(produtos2,
                  left_on = 'ProdutoID',  
                  right_on = 'ProdutoID',  
                  how = 'left') 

print("Número de linhas e colunas:",data1.shape)
data1.head()
```

```python
# Join com a tabela cabeçalho e funcionarios
cabecalho1 = cabecalho.merge(fun2,
                  left_on = 'FuncionarioID',   
                  right_on = 'FuncionarioID',  
                  how = 'left') 

# Join com a nova tabela cabeçalho e clientes
cabecalho2 = cabecalho1.merge(clientes,
                  left_on = 'ClienteID',   
                  right_on = 'ClienteID',  
                  how = 'left') 

print("Número de linhas e colunas:",cabecalho2.shape)
cabecalho2.head()
```

```python
# Join com a tabela data1 e cabeçalho
df = data1.merge(cabecalho2,
                  left_on = 'CupomID',   
                  right_on = 'CupomID',  
                  how = 'left') 

print("Número de linhas e colunas:",df.shape)
df.head()
```

#Tratamento do Data Frama

```python
# Verificando a existencia de nulo
round((df.isnull().sum()*100/len(df)).sort_values(ascending=False),2)
```

```python
# Verificando os tipos de variáveis
df.dtypes
```

```python
# Primeiramente vamos substituir , por .
df = df.replace({',': '.'}, regex=True)
```

```python
# Transformando algumas variáveis

datatypes_per_column = {
    'Valor':'float',
    'Desconto':'float',
    'Custo':'float',
    'ValorLiquido':'float',
    'ValorFrete':'float',
    'SalarioAnual':'float',
    'Data': 'datetime64[ms]',
    'DataEntrega': 'datetime64[ms]',
    'Extensao': 'object'
       }
  
df = df.astype(datatypes_per_column)
```

```python
# Verificando as alterações nos tipos de variáveis
df.dtypes
```

```python
df.to_csv('df.csv',index=False)
```
